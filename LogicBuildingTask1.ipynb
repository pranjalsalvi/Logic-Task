### 1. User Login Check

```python
# Task 1: User Login Check

username = "admin"
password = "1234"

# Input (you can replace with input() if needed)
input_username = "admin"
input_password = "1234"

# Check credentials
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### 2. Pass / Fail Analyzer

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Final result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

### 3. Simple Data Cleaner

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    clean_name = name.strip().lower()  # Remove spaces + convert to lowercase
    cleaned_names.append(clean_name)

print("Cleaned Names:", cleaned_names)
```

### 4. Message Length Analyzer

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")
    
    if length > 10:
        print("→ This message is longer than 10 characters")
```
