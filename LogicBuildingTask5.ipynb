```python
# Problem 1: Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example
analyze_engagement([200, 300, 150, 400])
```

```python
# Problem 2: Medicine Stock Alert
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

# Example
check_medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker
def rainfall_checker(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example
rainfall_checker([70, 80, 65, 75], 70)
```

```python
# Problem 4: Duplicate Account Detection
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(usernames) != len(unique_users):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

# Example
check_duplicates(["user1", "user2", "user1", "user3"])
```

```python
# Problem 5: Appointment Eligibility
def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

# Example
check_eligibility(21)
```

```python
# Problem 6: Premium Crop Price Filter
def premium_crops(prices):
    premium_list = []

    for price in prices:
        if price > 2000:
            premium_list.append(price)

    print("Premium Crops:", premium_list)

# Example
premium_crops([1500, 2500, 1800, 3200])
```

```python
# Problem 7: Application Health Checker
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

# Example
check_system_health(7)
```

```python
# Problem 8: Transaction Limit Checker
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

# Example
check_transaction(60000)
```
