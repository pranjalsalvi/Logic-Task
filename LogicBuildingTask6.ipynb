```python
# Problem 1: Smart Parking Lot Management System
def parking_system(capacity, logs):
    current = 0
    peak = 0

    # Process logs
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1

        # Track peak usage
        if current > peak:
            peak = current

    # Status check
    if current > capacity:
        status = "Full - Alert!"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)
    print("Parking Status:", status)

# Example
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem 2: Online Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    # Base time (5 min per km)
    time = distance * 5

    # Traffic impact
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    else:
        time += 5

    # Weather impact
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

# Example
delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_list):
    booked = len(booked_list)

    occupancy = (booked / total_seats) * 100

    # Status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)

# Example (simulate 150 bookings)
seat_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification
def server_load(cpu_readings):
    total = 0

    for cpu in cpu_readings:
        total += cpu

    avg = total / len(cpu_readings)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", avg, "%")
    print("Server Status:", status)

# Example
server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor
def resource_monitor(resources):
    overused = []

    # Threshold = 8 hours
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)

# Example
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# Problem 6: Online Event Registration Controller
def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

# Example
event_registration(100, 105)
```
